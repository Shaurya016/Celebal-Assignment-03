```python
# Step 1: Import pandas and load dataset from URL
import pandas as pd

url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)

# Step 2: Preview data
df.head()

```

```python
# Import visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Pairplot showing relationships
sns.set(style="whitegrid")
sns.pairplot(df, hue="species", corner=True)
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
# Boxplots for each feature
plt.figure(figsize=(14, 6))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(1, 4, i+1)
    sns.boxplot(x='species', y=column, data=df)
    plt.title(column)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.suptitle("Boxplots of Features by Species", y=1.05)
plt.show()
```

```python
# Heatmap of correlations
plt.figure(figsize=(8, 5))
sns.heatmap(df.iloc[:, :-1].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='species', data=df, palette='Set2')
plt.title("Number of Samples per Iris Species")
plt.show()
```

```python
species_counts = df['species'].value_counts()

plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Species Distribution")
plt.axis('equal')  # Equal aspect ratio makes the pie a circle
plt.show()
```
